```kotlin
import ds.utils.db.DB
import ds.utils.db.selectDf

var sirena = DB.selectDf("SELECT * FROM airlines.sirena_export")
```

```kotlin
import ds.types.custom.Document
import ds.types.custom.Name
import ds.utils.normalizeNameGOST

val rows = sirena
    .add("name") { Name.fromLFS(pax_name) }
//    .add("name_gost") { Name.fromLFS(pax_name, String::normalizeNameGOST) }
    .add("doc") { Document.from(travel_doc) }
    .mapToFrame {
        "id" from id
        "pax_name" from pax_name
		"first_name" from { "name"<Name>().first }    // common (not gost)
        "last_name" from { "name"<Name>().last }      //
        "second_name" from { "name"<Name>().second }  //
        "document" from { "doc"<Document>().toString() }
        "birth_date" from pax_birth_data
    }
```

```kotlin
import java.util.UUID

val users = rows
    .select { document and first_name and last_name and second_name and pax_name }
    .distinct()
    .add("id") { UUID.randomUUID() }
```

### Есть люди с одним паспортом на разеные имена

```kotlin
users.groupBy { document }
    .updateGroups {
        it.add("name") { "$last_name $first_name $second_name" }
            .select("name")
    }
    .aggregate { "name"<String>() into "names" }
    .filter { "names"<List<String>>().size > 1 }
```

```kotlin
import ds.utils.db.insertTo

users.insertTo("airlines.sirena_export_users")
```

```kotlin
val usersWithRenamedId = users.rename("id").into("user_id")
val usersRows = rows.join(usersWithRenamedId) { first_name and last_name and second_name and document }
```

```kotlin
import ds.utils.db.executeBatch

val unused = DB.getConnection().prepareStatement("UPDATE airlines.sirena_export SET user_id = ? WHERE id = ?;").use { ps ->
    usersRows.forEach {
        ps.setObject(1, it.user_id)
        ps.setObject(2, it.id)
        ps.addBatch()
    }
    ps.executeBatch()
}
```

```kotlin
import ds.utils.db.DB
import ds.utils.db.execute

DB.execute("""
UPDATE sirena_export_users usr
SET birth_date = t.bd
FROM (
    SELECT
        u.id AS id,
        se.pax_birth_data as bd
    FROM sirena_export se
    JOIN sirena_export_users u ON se.user_id = u.id
    WHERE se.pax_birth_data IS NOT NULL
) AS t
WHERE usr.id = t.id;
""")
```
