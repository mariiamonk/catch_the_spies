```kotlin
import ds.utils.db.DB
import ds.utils.db.selectDf

val usersDocs = DB.selectDf("select id, document from sirena_export_users")
```

```kotlin
import ds.types.custom.Document

val updated = usersDocs
    .update { document }
    	.with { Document.from(it).toNormalizedString() }
```

```kotlin
import ds.utils.db.execute

val unused = DB.execute("UPDATE sirena_export_users SET document = ? WHERE id = ?;") {
    updated.forEach {
        this@execute.setString(1, document)
        this@execute.setObject(2, id)
    	this@execute.addBatch()
    }
    executeBatch()
}
```
