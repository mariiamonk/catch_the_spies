```kotlin
%use dataframe(1.0.0-Beta3)
```

```kotlin
import ds.utils.db.DB
import ds.utils.db.execute
import ds.utils.db.selectTableAsDf
import ds.utils.db.insertTo
import java.util.UUID
```

```kotlin
val df = DB.selectTableAsDf("airlines.pointz_users")
```

```kotlin
val updated = df.update { first_name }.with { it.replace("'\''", "") }
    .update { last_name }.with { it.replace("'\'", "") }
    .select { id and first_name and last_name and uid }
    .convert { id }.to<UUID>()
```

```kotlin
updated.insertTo("wrk.pointz_normalized_users")
```

```kotlin
val df2 = DB.selectTableAsDf("airlines.forum_profiles")
```

```kotlin
val updatedProfiles = df2
    .update { first_name }.with { it?.replace("'\''", "") }
    .update { last_name }.with { it?.replace("'\'", "") }
    .select { id and first_name and last_name }
    .convert { id }.to<UUID>()
```

```kotlin
updatedProfiles.insertTo("wrk.forum_profiles_normalized")
```
