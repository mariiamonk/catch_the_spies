```kotlin
import ds.types.custom.Name
import java.util.UUID
import ds.utils.normalizeNameGOST
//import org.jetbrains.kotlinx.dataframe.api.to

val conv = sirenaUsers
    .rename("id").into("sirena_user_id")
    .convert { "sirena_user_id"() }.to<UUID>()
    .add {
        "name_v2" from { Name.fromLFS(pax_name, String::normalizeNameGOST) }
    }.add {
        "first_name_v2" from { "name_v2"<Name>().first }
        "last_name_v2" from { "name_v2"<Name>().last }
    }.remove { "name_v2"() }
    .update("document"<String>()).with {
        val serNo = it.split(" ").let { s -> s[0] to s[1] }
        "${serNo.first.padStart(4, '0')} ${serNo.second.padStart(6, '0')}"
    }
```

```kotlin
import ds.utils.db.executeBatch

var i = 0
val step = 100
val size = bdUsers.size().nrow

DB.getConnection().use { conn ->
    bdUsers.forEach {
        conn.prepareStatement("""
	INSERT INTO wrk.users (boarding_data_user_id, document, first_name_v2, last_name_v2)
		VALUES (?, ?, ?, ?)
		ON CONFLICT (document, first_name_v2, last_name_v2)
		DO UPDATE SET boarding_data_user_id = EXCLUDED.boarding_data_user_id;
        """).use { ps ->
            if (i++ % step == 0) {
                println(i)
            }
            ps.setObject(1, it.id)
            ps.setString(2, it.document)
            ps.setString(3, it.first_name)
            ps.setString(4, it.last_name)
            ps.executeUpdate()
        }
    }
}


//while (i + step < size) {
//    val unused = DB.getConnection().prepareStatement("""
//	INSERT INTO wrk.users (boarding_data_user_id, document, first_name_v2, last_name_v2)
//		VALUES (?, ?, ?, ?)
//		ON CONFLICT (document, first_name_v2, last_name_v2)
//		DO UPDATE SET boarding_data_user_id = EXCLUDED.boarding_data_user_id;
//	--	WHERE document = ? AND first_name_v2 = ? AND last_name_v2 = ?;
//
//	""").use { ps ->
//        bdUsers[i..<(i + step)].forEach {
//            ps.setObject(1, it.id)
//            ps.setString(2, it.document)
//            ps.setString(3, it.first_name)
//            ps.setString(4, it.last_name)
////        ps.setObject(5, it.id)
////        ps.setString(6, it.document)
////        ps.setString(7, it.first_name)
////        ps.setString(8, it.last_name)
//            ps.addBatch()
//        }
//        ps.executeBatch()
//    }
//
//}



```
