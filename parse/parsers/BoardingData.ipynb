```kotlin
import ds.types.raw.BoardingData
import ds.utils.db.inserter
import org.jetbrains.kotlinx.dataframe.api.convertTo

val ins = inserter<BoardingData>()

DataFrame.readCsv(
	fileOrUrl = "../data/BoardingData.csv",
	ignoreEmptyLines = true,
	delimiter = ';',
	parserOptions = ParserOptions(
		dateTimePattern = "MM/dd/yyyy",
	),
	colTypes = mapOf(
		"PassengerDocument" to ColType.String
	)
).renameToCamelCase()
	.convertTo <BoardingData> { }
	.toSequence()
	.apply { ins.insert(this) }

```
